**Finding optimum number of clusters**

```python
dist =[]

for i in range(1,11):
    kmeans = KMeans(n_clusters=i, random_state=0).fit(X)
    dist.append(kmeans.inertia_)
```

```python
fig, ax = plt.subplots(figsize=(10,6))

xplot = np.arange(1,11,1)
yplot = dist
plt.plot(xplot, yplot, linewidth=4)
plt.xlabel('Number of Clusters')
plt.ylabel('Within Cluster Sum of Squares')
plt.title('Elbow Method')
plt.show()    
```

From the Elbow Method, it is clear that 3 is the optimal number of clusters, since the Within Cluster Sum of Squares doesn't change much when Number of Clusters is increased above 3.

```python
pkmeans = KMeans(n_clusters=3, random_state=0)
fitted = pkmeans.fit_predict(X)
fitted
```

```python
Cluscent = pkmeans.cluster_centers_
Cluscent
```

```python
# Matching indices returned by KMeans.fit_predict() with y_train to identify corresponding name of Species
# rows 0,1,2 of mmatrix correspond to pkmeans value 0,1,2 respectively
# columns 0,1,2 of mmatrix correspond to Iris-setosa, Iris-versicolor, Iris-virginica respectively
# mmatrix consists of matches of flower species corresponding to pkmeans value

p=0
mmatrix = np.zeros([3,3])

for i in range(0,130):
    for j in range(0,3):
        if fitted[i]==j:
            if (y.iloc[i]=='Iris-setosa'): k=0
            elif (y.iloc[i]=='Iris-versicolor'): k=1
            elif (y.iloc[i]=='Iris-virginica'): k=2
                
            mmatrix[j,k]+=1
            
mmatrix
```

Based on the highest value in each row of the above matrix, it is clear that a fitted value of **0 indicates 'Iris-versicolor'**, fitted value of **1 indicates 'Iris-setosa'**,  and fitted value of **2 indicates 'Iris-virginica'**.

This matrix also shows the accuracy level of the Clustering performed above. Each row having 1 positive value and other two values as '0' corresponds to maximum accuracy, using the y_train vector as a benchmark.

```python
fig, ax = plt.subplots(figsize=(10,8))

plt.scatter(X.iloc[fitted==0, 0], X.iloc[fitted==0, 1], s=100, color='#6666ff', label='Iris-versicolor')
plt.scatter(X.iloc[fitted==1, 0], X.iloc[fitted==1, 1], s=100, color='#ff9933', label='Iris-setosa')
plt.scatter(X.iloc[fitted==2, 0], X.iloc[fitted==2, 1], s=100, color='#ff66cc', label='Iris-virginica')

plt.scatter(Cluscent[:,0], Cluscent[:,1], s=100, color='#000000', label='Cluster Center')

plt.legend()
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('Iris Clusters')
plt.show()
```
